```python
#shift+enter for executing current cell# 
from bs4 import BeautifulSoup as bs # beautifulsoup has been installed in cmd 
import requests
```

```python
link=("https://www.amazon.in/OnePlus-Nord-Jade-128GB-Storage/dp/B0B3CPQ5PF/ref=sr_1_12?crid=1W6JHL3J45JW0&keywords=oneplus+6t&qid=1674319783&sprefix=oneplus+6t%2Caps%2C288&sr=8-12")
```

```python
page=requests.get(link) #to get the link which has been pasted
```

```python
page
```

```python
page.content #to view content of webpage
```

```python
soup=bs(page.content,"html.parser")# to get it into html order format by taking soup as a variable#
```

```python
print(soup)
```

```python
names=soup.find_all('span',class_='a-profile-name')# to get customer names and tags(span)(this is the code snippet)#
#find_all is default function#
```

```python
names# got the names#
```

```python
customer_names=[]
for i in range(len(names)):
    customer_names.append(names[i].get_text())# to remove unneccessary tags in above names and get_text is default function
```

```python
customer_names
```

```python
customer_names.remove("Amazon Customer")# for removing amazon customer and also for any other duplicate/repetitive names
```

```python
customer_names# amazon customer removed and also we can use customer_name.pop(index for that name) to remove specific names
```

```python
reviews=soup.find_all('span',class_="a-size-base review-text")# this in bracket(class) is copied from URL by inspecting webpage#
# we need to find proper review having span and class 
```

```python
print(reviews)# got the review but unclear#
```

```python
cust_reviews=[]# this is copied from 35 above for getting only text and removing tags (cust=customer)
for i in range(len(reviews)):
    cust_reviews.append(reviews[i].get_text())
```

```python
print(cust_reviews)# we still have\n here and some unneccesary things
```

```python
cust_reviews=[i.strip("\n") for i in cust_reviews] # this is for removing \n at the above 
```

```python
cust_reviews # this all for just making one customer review printing neatly
```

```python
import pandas as pd # pandas is a module inbuilt 
```

```python
df=pd.DataFrame()
df["Name"]=customer_names
df["Reviews"]=cust_reviews
# these 3 lines are for creating a proper data frame in good order 
```
